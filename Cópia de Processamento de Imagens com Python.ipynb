Convertendo Imagens para arrays Numpy

```python
from matplotlib import image
from matplotlib import pyplot
# importando algumas bibliotecas da matplotlib, uteis para renderizar os dados do array numPy no formato imagem
```

Para carregar uma imagem como um array NumPy

```python
data = image.imread("sample_data/images/sao-paulo.jpg")
print(data)
# carregamos um arquivo de imagem e criamos uma vaariavel com o tipo Image, agora vamos carregar uma imagem do disco, como um array Numpy
# cada unidade temos 3 valores, que corresponde a tonalidade das cores
```

Imprimindo as propriedades do array de pixels

```python
# Imprimindo o tipo de dado
print('Tipo de dado:', data.dtype)
# Forma do array
print('Forma do array:', data.shape)
# tupla de 3 valores
# Valor máximo de pixels
print('Valor máximo de pixels:', data.max())
# Valor mínimo de pixels
print('Valor mínimo de pixels:', data.min())
```

### Exibindo o array de pixels como uma imagem

```python
# Ja carreguei minha imagem como um array e agora vou exibir esse array como imagem
pyplot.imshow(data)
# classe pyplot do matplotlib e o metodo imshow, como uma imagem digital, como se fosse um grafico normal
```

### Resize de Imagens com Python

Gerando thumbnail de tamanho 100x100

```python
image.thumbnail((100,100))
print('Dimensões da Imagem:', image.size)
image
# isso é importante quando estamos trabalhando com imagens aleatorias ou com poucos padrões
# Perceba que ele mantém a porcentagem de dimensão da imagem, para que a mesm não seja distorcida.
```

Carregando a imagem novamente

```python
image = Image.open("sample_data/images/sao-paulo.jpg")
# se quiser forçar isso, vamos utilizar o metodo resize
image_resize = image.resize((200,200))
print('Dimensões da Imagem:', image_resize.size)
pyplot.imshow(image_resize)
```

### Invertendo Imagens (Flip)

Outra operação muito util, principalmente quando estivermos trabalhando com metodos baseados em marching learning e deep learning, são a inversão de imagens. Muito util para gerar dados diferentes, para nosso conjunto de treinamento

Inversão horizontal

```python
horizontal_image = image.transpose(Image.FLIP_LEFT_RIGHT)
pyplot.imshow(horizontal_image)
# atributo FLIP
```

Inversão Vertical

```python
vertical_image = image.transpose(Image.FLIP_TOP_BOTTOM)
pyplot.imshow(vertical_image)
```
